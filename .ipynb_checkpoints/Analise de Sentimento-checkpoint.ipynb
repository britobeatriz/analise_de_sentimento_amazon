# Bibliotecas

```python
import pandas as pd
import seaborn as sns
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

## Importando os dados

```python
dados = pd.read_csv('./dados/dados_tratados.csv')
```

# Informações gerais dos dados

```python
print('Tamanho da base: ')
print(f'{dados.shape[0]} linhas e {dados.shape[1]} colunas.')
```

```python
print('Sentimentos presentes na base: ')
dados.sentimento.value_counts()
```

```python
print('Sentimentos em porcentagem: \n')
print((dados.sentimento.value_counts() / dados.shape[0]) * 100)
```

## Visualizando sentimentos no gráfico

```python
sns.catplot(x = 'sentimento', kind = 'count', data = dados, height = 5, aspect = 2)
```

A base está bastante desbalanceada e o sentimento de satisfação está bastante evidente.

## Nuvem de palavras

### Palavras comuns da base:

```python
values = ','.join(map(str, dados['texto']))
wordcloud = WordCloud(stopwords=values,
                      background_color="white",
                      width=1600, height=800).generate(values)

fig, ax = plt.subplots(figsize=(14,9))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
 
plt.imshow(wordcloud);
```

### Palavras mais utilizadas por pessoas que demonstraram satisfação:

```python
values=["satisfação"]
satisfacao = dados[dados.sentimento.isin(values)]
satisfacao
```

```python
values = ','.join(map(str, satisfacao['texto']))
wordcloud = WordCloud(stopwords=values,
                      background_color="white",
                      width=1600, height=800).generate(values)

fig, ax = plt.subplots(figsize=(14,9))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
 
plt.imshow(wordcloud);
```
